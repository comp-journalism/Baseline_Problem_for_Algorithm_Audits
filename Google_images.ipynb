### We need a way to identify unique images without looking through each image ourselves. 
#### The next script will read in each image, calculate a unique hash based on the image data, a bit like  a fingerprint, and store the hash along with the image filename in a shelf which is like a persistent dictionary. Lots of images will be repeats, so there will be lots of image filenames stored with the same hash key. 

#### The code is based on this blog post: https://realpython.com/blog/python/fingerprinting-images-for-near-duplicate-detection/ which I adapted to work with Python3

```python
%run index.py --dataset ./Data/clinton --shelve db.shelve
```

### Add the hash to each row in our `data` dataframe we loaded above

```python
import shelve

db = shelve.open('db.shelve')

data['image_hash'] = 0

for key in db.keys():   #  For every hash KEY
    for f in db[key]:   # for every file path name ITEM within the KEY
        for index, i in enumerate(data['image_path']):   # For every Image path in each row of my DF
            if f in i:                                   # If the ITEM file path is also in the IMAGE PATH of my DF
                data.image_hash[index] = key             # Put the KEY into the 'image_hash' Column

data.to_csv("./Data/hashedDF.csv", index=False)

db.close()
```
