# GoogleScraper 'Image Box' analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 3)
plt.rcParams['font.family'] = 'sans-serif'

pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
```

## Read in Google Scraper search results table 

```python
cols = ['id', 'link_type', 'link', 'visible_link', 'rank', 'serp_id','image_path']
```

```python
df = pd.read_csv('image_box/image_box_v3.csv', usecols=cols)
df.head()
```

## Read in Google Scraper serp table

```python
cols = ['id', 'requested_at', 'search_query']
serp = pd.read_csv('./image_box/serp_v3.csv', parse_dates=['requested_at'], usecols=cols)
serp.head()
```

```python
cols = ['serp_id', 'requested_at', 'search_query']
serp.columns = cols
```

```python
serp.head()
```

```python
type(serp.requested_at[0])
```

## Merge two together so we have dates associated with search results

```python
data = pd.merge(serp, df, on='serp_id')
```

```python
data.head()
```

```python
len(data)
```

## Create separate tables - one for each candidate

```python
HC = data[data.search_query == 'hillary clinton']
DT = data[data.search_query == 'donald trump']
BS = data[data.search_query == 'bernie sanders']
```

```python
HC['rank'].plot()
```

```python
DT['rank'].plot()
```

```python
BS['rank'].plot()
```

```python
HC.visible_link.value_counts()
```

```python
data.to_csv('rows_with_images.csv', index=False)
```
